```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('/kaggle/input/household/householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#Scatter plot with year aganist own
plt.scatter(data['income'],data['expenditure'])

#Adding title to the plot 
plt.title("Scatter plot")

#Setting the X and Y labels
plt.xlabel('income')
plt.ylabel('expenditure')

#Adding the Legends
plt.show()
```

```python
#Liine chart with year aganist own 
plt.plot(data['year'])
plt.plot(data['own'])

#Adding Title to the Plot
plt.title("Line Chart")

#Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('own')

#Adding the Legends
plt.show()
```

```python
#BAR CHART
plt.bar(data['year'],data['eqv_income'])

plt.title("Bar Chart")

#Setting the X and Y Labels
plt.xlabel('year')
plt.ylabel('eqv_income')

#Adding the Legends
plt.show()
```

```python
#HISTOGRAM
plt.hist(data['eqv_exp'])

plt.title("Histogram")

#Adding the Legends
plt.show()
```
