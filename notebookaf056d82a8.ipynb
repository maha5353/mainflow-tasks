```python
#creating a list
list=[1,36,53,63,96]

#adding a number
list.append(88)

#removing a number
list.remove(96)

#modifying a number
list[0]=538

#printing the updated list
print('Updated list: ',list)

```

```python
#creating a dictionary
dict={'Maharashtra':'Mumbai','Telangana':'Hyderabad','Karnataka':'Bangalore'}

#adding 
dict['Andhra pradesh']= 'Guntur'

#removing
del dict['Maharashtra']

#modifying
dict['Andhra pradesh']='Amaravathi'

#printing the updated list
print('Updated dictionary:',dict)
```
